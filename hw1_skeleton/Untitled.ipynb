```python
from sklearn.tree import DecisionTreeClassifier
import pandas as pd
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

# Dữ liệu từ bảng
data = {
    'ID': [1, 2, 3, 4, 5, 6, 7, 8],
    'Engine': ['2000cc', '1000cc', '2000cc', '1000cc', '1000cc', '2000cc', '1000cc', '2000cc'],
    'Type': ['SUV', 'Sedan', 'Sport', 'SUV', 'Sedan', 'Sport', 'Sedan', 'SUV'],
    'Color': ['Silver', 'Silver', 'Blue', 'Blue', 'Silver', 'Blue', 'Blue', 'Silver'],
    '4wd': ['Yes', 'Yes', 'No', 'No', 'Yes', 'Yes', 'No', 'No'],
    'Want': ['Yes', 'Yes', 'No', 'Yes', 'No', 'Yes', 'Yes', 'Yes']
}

# Chuyển dữ liệu thành DataFrame
df = pd.DataFrame(data)

# Chuyển các thuộc tính văn bản thành số
df['Engine'] = df['Engine'].map({'2000cc': 0, '1000cc': 1})
df['Type'] = df['Type'].map({'SUV': 0, 'Sedan': 1, 'Sport': 2})
df['Color'] = df['Color'].map({'Silver': 0, 'Blue': 1})
df['4wd'] = df['4wd'].map({'Yes': 0, 'No': 1})
df['Want'] = df['Want'].map({'Yes': 0, 'No': 1})

# Chia dữ liệu thành features và target
X = df.drop(['ID', 'Want'], axis=1)  # Features
y = df['Want']  # Target

# Tạo mô hình cây quyết định
model = DecisionTreeClassifier(random_state=42)
model.fit(X, y)

# Vẽ cây quyết định
plt.figure(figsize=(12, 8))
plot_tree(model, feature_names=X.columns, class_names=['No', 'Yes'], filled=True)
plt.show()
```
